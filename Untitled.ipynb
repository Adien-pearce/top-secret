```python
import pandas as pd
import warnings
import joblib
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Ignore warnings
warnings.filterwarnings('ignore')

# Load the dataset
try:
    data = pd.read_csv('collegePlace.csv')
except FileNotFoundError:
    print("Error: 'collegePlace.csv' not found. Please make sure the file is uploaded correctly.")

# --- Data Exploration and Preprocessing ---

# Display basic information about the data
print("Initial Data Information:")
print("Number of Rows:", data.shape[0])
print("Number of Columns:", data.shape[1])
print("\nFirst 5 rows:")
print(data.head())
print("\nData types and non-null values:")
print(data.info())
print("\nMissing values:")
print(data.isnull().sum())
print("\nDescriptive statistics:")
print(data.describe())
print("\nColumn names:")
print(data.columns)

# Map categorical columns to numerical values
data['Gender'].replace({'Male': 1, 'Female': 0}, inplace=True)
data['Stream'].replace({
    'Computer Science': 0,
    'Information Technology': 1,
    'Electronics And Communication': 2,
    'Mechanical': 3,
    'Electrical': 4,
    'Civil': 5
}, inplace=True)

# Define features (X) and target (y)
X = data.drop('PlacedOrNot', axis=1)
y = data['PlacedOrNot']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# --- Model Training ---

# Initialize and train the models
lr = LogisticRegression()
lr.fit(X_train, y_train)

svm_model = svm.SVC()
svm_model.fit(X_train, y_train)

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)

# --- Model Evaluation ---

# Make predictions
y_pred1 = lr.predict(X_test)
y_pred2 = svm_model.predict(X_test)
y_pred3 = knn.predict(X_test)
y_pred4 = dt.predict(X_test)
y_pred5 = rf.predict(X_test)
y_pred6 = gb.predict(X_test)

# Calculate accuracy scores
score1 = accuracy_score(y_test, y_pred1)
score2 = accuracy_score(y_test, y_pred2)
score3 = accuracy_score(y_test, y_pred3)
score4 = accuracy_score(y_test, y_pred4)
score5 = accuracy_score(y_test, y_pred5)
score6 = accuracy_score(y_test, y_pred6)

print("\nAccuracy Scores:")
print(f"Logistic Regression: {score1}")
print(f"SVM: {score2}")
print(f"K-Nearest Neighbors: {score3}")
print(f"Decision Tree: {score4}")
print(f"Random Forest: {score5}")
print(f"Gradient Boosting: {score6}")

# Create a DataFrame for visualization
final_data = pd.DataFrame({
    'Models': ['LR', 'SVC', 'KNN', 'DT', 'RF', 'GB'],
    'ACC': [score1 * 100, score2 * 100, score3 * 100, score4 * 100, score5 * 100, score6 * 100]
})

final_data = final_data.sort_values(by='ACC', ascending=False)
print("\nFinal Data (Sorted by Accuracy):")
print(final_data)

# --- Visualization ---

plt.figure(figsize=(10, 6))
sns.barplot(x=final_data['Models'], y=final_data['ACC'], palette='viridis')
plt.title('Model Accuracy Comparison', fontsize=16)
plt.xlabel('Models', fontsize=12)
plt.ylabel('Accuracy (%)', fontsize=12)
plt.xticks(rotation=45)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('model_accuracy_bar_plot.png')
print("Saved bar plot to 'model_accuracy_bar_plot.png'")

# --- Sample Prediction ---

new_data = pd.DataFrame({
    'Age': [22],
    'Gender': [1],
    'Stream': [0],
    'Internships': [1],
    'CGPA': [8],
    'Hostel': [1],
    'HistoryOfBacklogs': [0]
})

lr_full = LogisticRegression()
lr_full.fit(X, y)

p = lr_full.predict(new_data)
prob = lr_full.predict_proba(new_data)

print("\nSample Prediction:")
if p[0] == 1:
    print('Placed')
    print(f"You will be placed with probability of {prob[0][1]:.2f}")
else:
    print("Not Placed")

# --- Model Saving and Loading ---

joblib.dump(lr_full, 'model_college_placement.joblib')
print("Saved Logistic Regression model to 'model_college_placement.joblib'")

model_loaded = joblib.load('model_college_placement.joblib')
print("\nLoaded the model from 'model_college_placement.joblib'")
print("Loaded model prediction for new data:", model_loaded.predict(new_data))
```
